# Incarico e Obiettivo

Il Ministero Economie e Finanze tramite consip, acquisisce e pubblica tutti gli acquisti fatti dai vari enti della Pubblica Amministrazione. I dati (scaricabili dal sito dati.consip.it) sono organizzati in:
- Amministrazioni del programma: Pubbliche Amministrazioni registrate al Programma di razionalizzazione degli acquisti
- Bandi e Gare: Iniziative di gara e lotti pubblicati nell'ambito del Programma di razionalizzazione degli acquisti
- Fornitori: aziende partecipanti al programma
- Cataloghi: Beni, Servizi e Lavori disponibili tramite le iniziative del Programma di razionalizzazione degli acquisti
- Negoziazioni: Negoziazioni telematiche effettuate dalle PA abilitate finalizzate alla selezione di offerte per l'acquisto di beni e servizi
- Acquisti: Acquisti formalizzati tra PA e Fornitori nell'ambito del Programma di razionalizzazione

Questi dati servono a tracciare tutti gli acquisti di servizi, beni fatti dalla pubblica amministrazione verso le aziende iscritte al registro pubblico delle aziende.

L’obiettivo di questo esercizio e’ costruire una exploratory data analysis ed un modello predittivo partendo da questi dati. La traccia è’ volutamente libera, e permetterà al candidato di mostrare le proprie capacità propositive sia nell’individuare domande di ricerca che nel valutarle. 
Il candidato, potrà integrare questi dati con altre sorgenti dati come: indice pa e/o ANAC.

# Domanda di ricerca

In questa sezione del notebook esporremo la nostra domanda di ricerca insieme ad una sintesi dei requisiti e delle considerazioni a livello di business.
### Introduzione
Il valore economico totale di ogni acquisto effettuato dalla PA dipende da numerose variabili tra cui il numero di ordini, il tipo di bene o servizio ordinato, la regione dove risiede la PA, ecc. Poter prevedere il valore economico di ogni acquisto potrebbe permettere alle varie amministrazioni di prevedere la spesa da affrontare in anticipo e quindi di poter pianificare e programmare il proprio budget in modo più efficiente. 
In questa analisi, proviamo a costruire un modello che sia in grado di prevedere il valore economico totale di una spesa considerando diverse variabili come: la regione del fornitore, la regione della pubblica amministrazione, il numero fornitori, il numero ordini, la tipologia di amministrazione, l'inflazione annuale media, il tipo di bene e/o servizio acquistato, ecc.

### Domanda di ricerca 
                      "È possibile prevedere il valore economico di un acquisto?"

### Inflazione
#### in questa sezione aggiungiamo al nostro dataset il tasso di inflazione media in Italia (CPI) per l'anno di riferimento

```python
# Definiamo una funzione generale per agguingere il tasso d'inflazione medio dell'anno di riferimento; source: https://it.inflation.eu/tassi-di-inflazione/italia/inflazione-storica/cpi-inflazione-italia.aspx
def inflazione(row):
    if row['Anno_Riferimento'] == 2018:
        return '1.14'
    if row['Anno_Riferimento'] == 2019:
        return '0.61'
    if row['Anno_Riferimento'] == 2020:
        return '0.29'
    
df_finale['Inflazione'] = df_finale.apply(lambda row: inflazione(row), axis=1)
```

```python
#random sample
df_finale = df_finale.sample(frac=0.1, random_state=1)


# sklearn one hot encoding: mappatura di ogni categoria come 0 (cold) o 1 (hot) 
# creare ndarray per one hot encodoing (sklearn)
tipologia_amministrazione = df_finale.iloc[:,1:2].values 
ohe = OneHotEncoder() 
tipologia_amministrazione = ohe.fit_transform(tipologia_amministrazione).toarray()
tipologia_amministrazione = pd.DataFrame(tipologia_amministrazione)
tipologia_amministrazione.columns = df_finale.Tipologia_Amministrazione.unique()

regione_pa = df_finale.iloc[:,2:3].values 
ohe = OneHotEncoder() 
regione_pa = ohe.fit_transform(regione_pa).toarray()
regione_pa = pd.DataFrame(regione_pa)
regione_pa.columns = df_finale.Regione_PA.unique()

regione_fornitore = df_finale.iloc[:,3:4].values 
ohe = OneHotEncoder() 
regione_fornitore = ohe.fit_transform(regione_fornitore).toarray()
regione_fornitore = pd.DataFrame(regione_fornitore)
regione_fornitore.columns = df_finale.Regione_Fornitore.unique()

accordo_quadro = df_finale.iloc[:,4:5].values 
ohe = OneHotEncoder() 
accordo_quadro = ohe.fit_transform(accordo_quadro).toarray()
accordo_quadro = pd.DataFrame(accordo_quadro)
accordo_quadro.columns = df_finale.Accordo_Quadro.unique()

lotto = df_finale.iloc[:,5:6].values 
ohe = OneHotEncoder() 
lotto = ohe.fit_transform(lotto).toarray()
lotto = pd.DataFrame(lotto)
lotto.columns = df_finale.Lotto.unique()

bene_servizio = df_finale.iloc[:,6:7].values 
ohe = OneHotEncoder() 
bene_servizio = ohe.fit_transform(bene_servizio).toarray()
bene_servizio = pd.DataFrame(bene_servizio)
bene_servizio.columns = df_finale.Bene_Servizio.unique()


tipologia_acquisto = df_finale.iloc[:,14:15].values 
ohe = OneHotEncoder() 
tipologia_acquisto = ohe.fit_transform(tipologia_acquisto).toarray()
tipologia_acquisto = pd.DataFrame(tipologia_acquisto)
tipologia_acquisto.columns = df_finale.Tipologia_Acquisto.unique()


codice_cpv = df_finale.iloc[:,7:8].values 
ohe = OneHotEncoder() 
codice_cpv = ohe.fit_transform(codice_cpv).toarray()
codice_cpv = pd.DataFrame(codice_cpv)
codice_cpv.columns = df_finale.Codice_CPV.unique()


strumento_acquisto = df_finale.iloc[:,15:16].values 
ohe = OneHotEncoder() 
strumento_acquisto = ohe.fit_transform(strumento_acquisto).toarray()
strumento_acquisto = pd.DataFrame(strumento_acquisto)
strumento_acquisto.columns = df_finale.Strumento_Acquisto.unique()


bando_mepa = df_finale.iloc[:,16:17].values 
ohe = OneHotEncoder() 
bando_mepa = ohe.fit_transform(bando_mepa).toarray()
bando_mepa = pd.DataFrame(bando_mepa)
bando_mepa.columns = df_finale.Bando_Mepa.unique()


categoria_abilitazione = df_finale.iloc[:,17:18].values 
ohe = OneHotEncoder() 
categoria_abilitazione = ohe.fit_transform(categoria_abilitazione).toarray()
categoria_abilitazione = pd.DataFrame(categoria_abilitazione)
categoria_abilitazione.columns = df_finale.Categoria_Abilitazione.unique()


convenzione = df_finale.iloc[:,19:20].values 
ohe = OneHotEncoder() 
convenzione = ohe.fit_transform(convenzione).toarray()
convenzione = pd.DataFrame(convenzione)
convenzione.columns = df_finale.Convenzione.unique()


bando_sda = df_finale.iloc[:,23:24].values 
ohe = OneHotEncoder() 
bando_sda = ohe.fit_transform(bando_sda).toarray()
bando_sda = pd.DataFrame(bando_sda)
bando_sda.columns = df_finale.Bando_SDA.unique()


aq = df_finale.iloc[:,27:28].values 
ohe = OneHotEncoder() 
aq = ohe.fit_transform(aq).toarray()
aq = pd.DataFrame(aq)
aq.columns = df_finale.AQ.unique()


tipo_negoziazione = df_finale.iloc[:,28:29].values 
ohe = OneHotEncoder() 
tipo_negoziazione = ohe.fit_transform(tipo_negoziazione).toarray()
tipo_negoziazione = pd.DataFrame(tipo_negoziazione)
tipo_negoziazione.columns = df_finale.Tipo_negoziazione.unique()


#take the numerical data from the original data
X_num = df_finale[['N_Ordini', 'N_PA', 'N_PO', 'N_Fornitori', 'N_Ordini_Consumi', 'N_PA_Consumi'
             , 'N_PO_Consumi', 'Durata_Media_Fornitura', 'N_AS_Pubblicati', 'N_PA_Appaltanti', 
             'N_Negoziazioni_Pubblicate', 'N_Medio_Fornitori_Invitati', 'Tempo_Medio_Offerte', 'Inflazione']].copy()


#reset index to avoid error
# df_finale.reset_index(drop=True, inplace=True)
X_num.reset_index(drop=True, inplace=True)
regione_fornitore.reset_index(drop=True, inplace=True)
accordo_quadro.reset_index(drop=True, inplace=True) 
lotto.reset_index(drop=True, inplace=True)
bene_servizio.reset_index(drop=True, inplace=True)
codice_cpv.reset_index(drop=True, inplace=True)
tipologia_acquisto.reset_index(drop=True, inplace=True)
strumento_acquisto.reset_index(drop=True, inplace=True)
bando_mepa.reset_index(drop=True, inplace=True)
categoria_abilitazione.reset_index(drop=True, inplace=True)
convenzione.reset_index(drop=True, inplace=True)
bando_sda.reset_index(drop=True, inplace=True)
aq.reset_index(drop=True, inplace=True)
tipo_negoziazione.reset_index(drop=True, inplace=True)
regione_pa.reset_index(drop=True, inplace=True)
tipologia_amministrazione.reset_index(drop=True, inplace=True)



#take the encoded data and add to numerical data
X_final = pd.concat([X_num, regione_fornitore, accordo_quadro, lotto, bene_servizio, codice_cpv, 
                    tipologia_acquisto, strumento_acquisto, bando_mepa, categoria_abilitazione, convenzione,
                    bando_sda, aq, tipo_negoziazione, regione_pa, tipologia_amministrazione], axis = 1)
X_final.fillna(0, inplace=True)

#define y as being the "valore economico column" from the original dataset
y_final = df_finale[['Valore_Economico']].copy()

#Test train split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.20, random_state = 0)
#Standardization
sc = StandardScaler()
X_train = sc.fit_transform(X_train.astype(np.float))
X_test= sc.transform(X_test.astype(np.float))


```

#### Random Forest 
random forest is based on ensemble learning (multiple learning methods are used) predicted value is calculated from the mean of all outputs. Through bagging, we divide the dataset into smaller samples and creates multiple trees.
Every tree as a result of the division are shallower. 
Overfitting is reduced compare to simple decision tree due to aggregations. 
It does not need scaling and it works well with outliers.

```python
forest = RandomForestRegressor(n_estimators = 100,
                              criterion = 'mse',
                              random_state = 1,
                              n_jobs = -1,
                               max_depth = 5
                              )


#fit model (ravel faster since no memory is copied)
forest.fit(X_train, y_train.values.ravel())
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)

# Evaluating the Random forest Regression
print('Mean Absolute Error: %f' % (metrics.mean_absolute_error(y_test, y_test_pred)))  
print('Mean Squared Error: %f' %(metrics.mean_squared_error(y_test, y_test_pred)))
print('Root Mean Squared Error: %f' %(np.sqrt(metrics.mean_squared_error(y_test, y_test_pred))))
```

```python
### print score
print('forest train R squared score %.3f, forest R squared test score: %.3f' % (
forest.score(X_train,y_train),
forest.score(X_test, y_test)))
```
